```python
# ------------------------------------------------------------
# 2.1 Calculate clustering scores
# ------------------------------------------------------------

# This function returns the main clustering quality scores.
# For DBSCAN and OPTICS, label -1 means noise.

def calculate_clustering_scores(X, cluster_labels, ignore_noise=True, sample_size=2000):

    labels = np.asarray(cluster_labels)

    # Remove noise labels when scoring DBSCAN / OPTICS clusters.
    
    if ignore_noise:
        keep_mask = labels != -1
        X_for_score = X.loc[keep_mask] if hasattr(X, "loc") else X[keep_mask]
        labels_for_score = labels[keep_mask]
    else:
        X_for_score = X
        labels_for_score = labels

    unique_labels = np.unique(labels_for_score)

    # Scores are valid only when at least 2 real clusters exist.
    
    if len(unique_labels) < 2 or len(unique_labels) >= len(labels_for_score):
        return {
            "Silhouette": np.nan,
            "Davies_Bouldin": np.nan,
            "Calinski_Harabasz": np.nan
        }

    # Use sampling for Silhouette if the dataset is large, so the notebook stays fast.
    
    if len(labels_for_score) > sample_size:
        silhouette_value = silhouette_score(
            X_for_score,
            labels_for_score,
            sample_size=sample_size,
            random_state=42
        )
    else:
        silhouette_value = silhouette_score(X_for_score, labels_for_score)

    return {
        "Silhouette": silhouette_value,
        "Davies_Bouldin": davies_bouldin_score(X_for_score, labels_for_score),
        "Calinski_Harabasz": calinski_harabasz_score(X_for_score, labels_for_score)
    }
```

```python
# ------------------------------------------------------------
# 2.2 Create business profile for cluster labels
# ------------------------------------------------------------

# This function converts cluster numbers into a readable business summary.

def create_cluster_profile(cluster_labels, label_name):

    # Keep the original RFM values because they are easier to interpret.
    
    profile_data = rfm[[
        "CustomerID",
        "Recency",
        "Frequency",
        "Monetary",
        "Country"
    ]].copy()

    # Add the cluster labels created by the selected method.
    
    profile_data[label_name] = cluster_labels

    # Summarise customer behaviour inside each cluster.
    profile = (
        profile_data
        .groupby(label_name)
        .agg(
            Customers=("CustomerID", "count"),
            Avg_Recency=("Recency", "mean"),
            Avg_Frequency=("Frequency", "mean"),
            Avg_Monetary=("Monetary", "mean"),
            Total_Monetary=("Monetary", "sum")
        )
        .reset_index()
    )

    # Add share columns for easier business interpretation.
    profile["Customer_Share_%"] = profile["Customers"] / len(profile_data) * 100
    profile["Revenue_Share_%"] = profile["Total_Monetary"] / profile_data["Monetary"].sum() * 100

    return profile.round(2)
```

### DBSCAN settings

DBSCAN mainly needs two settings.

| Setting | Simple meaning |
|---|---|
| `eps` | How close customers must be to count as neighbours |
| `min_samples` | How many nearby customers are needed to form a dense group |

Think of `eps` as a circle around one customer.

If enough customers are inside that circle, DBSCAN calls it a dense area.

If not, the customer may become noise.

```python
# Purpose: run this notebook step using the prepared customer data.
# ------------------------------------------------------------
# 5.1 Calculate k-distance values
# ------------------------------------------------------------
# We use n_neighbors=5 because min_samples=5 is a common DBSCAN starting point.

neighbors = NearestNeighbors(n_neighbors=5)
neighbors.fit(X_scaled)

distances, indices = neighbors.kneighbors(X_scaled)

k_distance = np.sort(distances[:, -1])

kdist_p90 = np.percentile(k_distance, 90)
kdist_p95 = np.percentile(k_distance, 95)
kdist_p98 = np.percentile(k_distance, 98)

print(f"90th percentile distance: {kdist_p90:.4f}")
print(f"95th percentile distance: {kdist_p95:.4f}")
print(f"98th percentile distance: {kdist_p98:.4f}")
```

  <table style="border-collapse: collapse; width: 100%; margin: 12px 0;">

    <code>eps = 0.5, 1.0, 2.0</code><br>
    <code>min_samples = 3, 5, 10</code>
  </div>

  <p>
    For each combination, we check three things:
  </p>

  <ul>
    <li>How many clusters DBSCAN created</li>
    <li>How many customers became noise</li>
    <li>How good the clustering scores are</li>
  </ul>

  <div style="
    border-left: 4px solid #22c55e;
    padding: 10px 14px;
    margin: 12px 0;
    background: var(--vscode-editorWidget-background);
    border-radius: 6px;
  ">
    <strong>Goal of this step:</strong><br>
    Find a practical DBSCAN setting that creates useful customer groups without marking too many customers as noise.
  </div>

</div>

```python
# ------------------------------------------------------------
# 6.1 Run DBSCAN grid search
# ------------------------------------------------------------

eps_values = [0.5, 1.0, 2.0]
min_samples_values = [3, 5, 10]

dbscan_records = []
dbscan_labels_store = {}

for eps in eps_values:
    for min_samples in min_samples_values:

        # Run DBSCAN for one eps + min_samples combination.
        dbscan_model = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan_model.fit_predict(X_scaled)

        # Count real clusters and noise customers.
        cluster_count = len(set(labels) - {-1})
        noise_count = int((labels == -1).sum())
        noise_share = noise_count / len(labels) * 100

        # Calculate quality scores for valid clusters.
        metrics = calculate_clustering_scores(
            X_scaled,
            labels,
            ignore_noise=True
        )

        # Store result for this setting.
        dbscan_records.append({
            "eps": eps,
            "min_samples": min_samples,
            "Clusters": cluster_count,
            "Noise_Customers": noise_count,
            "Noise_%": noise_share,
            **metrics
        })

        # Store labels so we can reuse the best setting later.
        dbscan_labels_store[(eps, min_samples)] = labels

dbscan_grid_results = pd.DataFrame(dbscan_records).round(4)

dbscan_grid_results.to_csv(
    OUTPUT_DIR / "dbscan_grid_results.csv",
    index=False
)

dbscan_grid_results
```

    background: var(--vscode-editorWidget-background);
    border-radius: 6px;
  ">
    <strong>Goal of this step:</strong><br>
    Add a density score so we can understand which customers sit in common areas
    and which customers are less typical.
  </div>

</div>

```python
# ------------------------------------------------------------
# 10.1 Create density scores
# ------------------------------------------------------------

# Higher density score = customer is in a more common behaviour area.

kde_model = KernelDensity(
    kernel="gaussian",
    bandwidth=0.5
)

# Fit KDE on scaled RFM values.

kde_model.fit(X_scaled)

# Convert KDE log-scores into normal density scores.

density_scores = np.exp(kde_model.score_samples(X_scaled))

# Mark top 10% density customers as high-density region.

density_threshold = np.percentile(density_scores, 90)

# Keep original RFM values for business interpretation.

density_table = rfm[[
    "CustomerID",
    "Recency",
    "Frequency",
    "Monetary",
    "Country"
]].copy()

density_table["Density_Score"] = density_scores
density_table["High_Density_Region"] = density_table["Density_Score"] >= density_threshold

# Summarise customer behaviour by density group.

density_summary = (
    density_table
    .groupby("High_Density_Region")
    .agg(
        Customers=("CustomerID", "count"),
        Avg_Recency=("Recency", "mean"),
        Avg_Frequency=("Frequency", "mean"),
        Avg_Monetary=("Monetary", "mean")
    )
    .reset_index()
    .round(2)
)

density_table.to_csv(
    OUTPUT_DIR / "denclue_style_density_scores.csv",
    index=False
)

density_summary
```

```python
# ------------------------------------------------------------
# 11.1 Create grid summary
# ------------------------------------------------------------
# Goal:
# Divide the PCA customer map into grid cells and find dense customer areas.

# Add original RFM values to the PCA map.
# PCA values show customer position.
# RFM values help explain business behaviour inside each grid cell.
grid_data = pca_view.merge(
    rfm[["CustomerID", "Recency", "Frequency", "Monetary"]],
    on="CustomerID",
    how="left"
)

# Split the PC1 axis into 10 equal-sized grid bins.
grid_data["PC1_Bin"] = pd.cut(
    grid_data["PC1"],
    bins=10,
    labels=False,
    duplicates="drop"
)

# Split the PC2 axis into 10 equal-sized grid bins.
grid_data["PC2_Bin"] = pd.cut(
    grid_data["PC2"],
    bins=10,
    labels=False,
    duplicates="drop"
)

# Summarise customers inside each grid cell.
grid_summary = (
    grid_data
    .groupby(["PC1_Bin", "PC2_Bin"])
    .agg(
        Customers=("CustomerID", "count"),
        Avg_Recency=("Recency", "mean"),
        Avg_Frequency=("Frequency", "mean"),
        Avg_Monetary=("Monetary", "mean"),
        Total_Monetary=("Monetary", "sum")
    )
    .reset_index()
)

# Create a readable cell ID like "3_5".
grid_summary["Cell_ID"] = (
    grid_summary["PC1_Bin"].astype(str)
    + "_"
    + grid_summary["PC2_Bin"].astype(str)
)

# Dense cell rule:
# A cell is dense if it has customer count above the 75th percentile,
# but the cutoff must be at least 10 customers.
dense_cell_cutoff = max(
    10,
    int(np.percentile(grid_summary["Customers"], 75))
)

# Mark grid cells as dense or not dense.
grid_summary["Dense_Cell"] = grid_summary["Customers"] >= dense_cell_cutoff

# Show the busiest cells first.
grid_summary = (
    grid_summary
    .sort_values("Customers", ascending=False)
    .round(2)
)

# Save grid summary for documentation.
grid_summary.to_csv(
    OUTPUT_DIR / "grid_sting_summary.csv",
    index=False
)

print("Dense cell cutoff:", dense_cell_cutoff)
print("Dense cells:", int(grid_summary["Dense_Cell"].sum()))

display(grid_summary.head(10))
```

```python
# ------------------------------------------------------------
# 12.1 Create WaveCluster-style smoothed density grid
# ------------------------------------------------------------
# Goal:
# Find broader dense customer regions on the PCA map.

# Step 1:
# Put a 25 x 25 grid over the PCA customer map.
# Each cell counts how many customers fall inside that area.

customer_grid, pc1_edges, pc2_edges = np.histogram2d(
    pca_view["PC1"],
    pca_view["PC2"],
    bins=25
)

# Step 2:
# Smooth the grid.
# This combines nearby busy cells into broader dense regions.

smoothed_grid = gaussian_filter(
    customer_grid,
    sigma=1.2
)

# Step 3:
# Define high-density cells as the top 10% of non-empty smoothed cells.

density_cutoff = np.percentile(
    smoothed_grid[smoothed_grid > 0],
    90
)

# Step 4:
# Find the grid positions where density is high.

high_density_cells = np.argwhere(smoothed_grid >= density_cutoff)

# Step 5:
# Create a readable summary table.

wave_summary = pd.DataFrame(
    high_density_cells,
    columns=["PC1_Grid_Index", "PC2_Grid_Index"]
)

# Step 6:
# Add the density value for each high-density cell.

wave_summary["Smoothed_Density"] = [
    smoothed_grid[pc1_index, pc2_index]
    for pc1_index, pc2_index in high_density_cells
]

# Step 7:
# Sort strongest dense areas first.

wave_summary = (
    wave_summary
    .sort_values("Smoothed_Density", ascending=False)
    .round(4)
)

# Save result for documentation.

wave_summary.to_csv(
    OUTPUT_DIR / "wavecluster_style_summary.csv",
    index=False
)

print("High smoothed-density cells:", len(wave_summary))

display(wave_summary.head(10))
```

  <h3 style="margin-bottom: 6px;">13. CLIQUE-style subspace check</h3>

We check:

Recency + Frequency

Recency + Monetary

Frequency + Monetary

```python
# ------------------------------------------------------------
# 13.1 Check dense cells in feature pairs
# ------------------------------------------------------------
# Goal:

# Check which two-feature view shows strong customer concentration.

# We check three smaller views of the scaled RFM data.

subspace_pairs = [
    ("Recency_Scaled", "Frequency_Scaled"),
    ("Recency_Scaled", "Monetary_Scaled"),
    ("Frequency_Scaled", "Monetary_Scaled")
]

subspace_records = []

for feature_1, feature_2 in subspace_pairs:

    # Keep only CustomerID and the current two features.
    
    pair_data = rfm_scaled[["CustomerID", feature_1, feature_2]].copy()

    # Put feature_1 values into 10 groups from low to high.
    
    pair_data["Feature_1_Bin"] = pd.cut(
        pair_data[feature_1],
        bins=10,
        labels=False,
        duplicates="drop"
    )

    # Put feature_2 values into 10 groups from low to high.
    
    pair_data["Feature_2_Bin"] = pd.cut(
        pair_data[feature_2],
        bins=10,
        labels=False,
        duplicates="drop"
    )

    # Count customers inside each two-feature box.
    
    cell_counts = (
        pair_data
        .groupby(["Feature_1_Bin", "Feature_2_Bin"])
        .size()
        .reset_index(name="Customers")
    )

    # A dense cell must have at least 1% of customers.
    # For 4,312 customers, this is about 43 customers.
    
    density_cutoff = max(
        10,
        int(len(pair_data) * 0.01)
    )

    # Keep only boxes with enough customers.
    
    dense_cells = cell_counts[
        cell_counts["Customers"] >= density_cutoff
    ]

    # Store one summary row for this feature pair.
    
    subspace_records.append({
        "Subspace": f"{feature_1} + {feature_2}",
        "Dense_Cells": len(dense_cells),
        "Largest_Cell_Customers": int(cell_counts["Customers"].max()),
        "Density_Cutoff": density_cutoff
    })

# Final summary table.

clique_summary = pd.DataFrame(subspace_records)

clique_summary.to_csv(
    OUTPUT_DIR / "clique_style_subspace_summary.csv",
    index=False
)

clique_summary
```

  <h3 style="margin-bottom: 6px;">14. Constraint-based business segments</h3>

  <p>
    Machine learning models find patterns in the data.
    But they do not automatically understand business priorities.
  </p>

  <p>
    For example, a model may place a very high-value customer inside a normal cluster.
    But from a business point of view, that customer may need VIP treatment.
  </p>

  <table style="border-collapse: collapse; width: 100%; margin: 12px 0;">
    <thead>
      <tr>
        <th style="text-align: left; padding: 8px; border-bottom: 1px solid var(--vscode-panel-border);">Business rule</th>
        <th style="text-align: left; padding: 8px; border-bottom: 1px solid var(--vscode-panel-border);">Why it matters</th>
      </tr>
    </thead>
    <tbody>
      <tr>
        <td style="padding: 8px; border-bottom: 1px solid var(--vscode-panel-border);">
          Very high-value customers should be treated as VIP.
        </td>
        <td style="padding: 8px; border-bottom: 1px solid var(--vscode-panel-border);">
          They may contribute a large share of revenue and need special attention.
        </td>
      </tr>
      <tr>
        <td style="padding: 8px; border-bottom: 1px solid var(--vscode-panel-border);">
          Very inactive low-frequency customers should be reactivation targets.
        </td>
        <td style="padding: 8px; border-bottom: 1px solid var(--vscode-panel-border);">
          They may need win-back campaigns instead of normal marketing.
        </td>
      </tr>
    </tbody>
  </table>

  <p>
    So we add a simple business-rule layer on top of the K-Means baseline.
  </p>

  <div style="
    border-left: 4px solid #38bdf8;
    padding: 10px 14px;
    margin: 12px 0;
    background: var(--vscode-editorWidget-background);
    border-radius: 6px;
  ">
    <strong>Simple idea:</strong><br>
    K-Means gives the base customer groups.
    Business rules adjust the final label when a customer clearly needs special treatment.
  </div>

  <div style="
    border-left: 4px solid #22c55e;
    padding: 10px 14px;
    margin: 12px 0;
    background: var(--vscode-editorWidget-background);
    border-radius: 6px;
  ">
    <strong>Goal of this step:</strong><br>
    Make the segmentation more useful for business action by adding rule-based labels such as VIP and reactivation target.
  </div>

</div>

| Step                                    | Meaning                                                                               |
| --------------------------------------- | ------------------------------------------------------------------------------------- |
| Start with `KMeans_0`, `KMeans_1`, etc. | Every customer first keeps their K-Means segment                                      |
| Rule 1                                  | Very high-value or strong high-value outlier customers become `Business_VIP_Outlier`  |
| Rule 2                                  | Inactive customers with low buying frequency become `Business_Inactive_Low_Frequency` |
| Final summary                           | Count customers and revenue share per final segment                                   |


```python
# ------------------------------------------------------------
# 14.1 Create constraint-based business segments
# ------------------------------------------------------------
# Goal:
# Start with K-Means clusters, then add business rules for special customer groups.

# Keep original customer behaviour columns.

constraint_data = rfm[[
    "CustomerID",
    "Recency",
    "Frequency",
    "Monetary",
    "Country"
]].copy()

# Add K-Means cluster labels from Notebook 2.

constraint_data = constraint_data.merge(
    customer_partitioning[["CustomerID", "KMeans_Cluster"]],
    on="CustomerID",
    how="left"
)

# Add strong outlier flag from Notebook 2.

constraint_data = constraint_data.merge(
    outlier_detection[["CustomerID", "Strong_Outlier_Flag"]],
    on="CustomerID",
    how="left"
)

# Default segment:
# Every customer first gets their K-Means cluster name.

constraint_data["Constraint_Segment"] = (
    "KMeans_"
    + constraint_data["KMeans_Cluster"].astype(str)
)

# Business rule cutoffs.
vip_cutoff = constraint_data["Monetary"].quantile(0.99)
high_value_cutoff = constraint_data["Monetary"].quantile(0.95)
inactive_cutoff = constraint_data["Recency"].quantile(0.75)

# Rule 1:
# Mark VIP customers.
# A customer becomes VIP if:
# - they are in the top 1% by Monetary value, OR
# - they are a strong outlier and also in the top 5% by Monetary value.

constraint_data.loc[
    (constraint_data["Monetary"] >= vip_cutoff)
    | (
        (constraint_data["Strong_Outlier_Flag"] == True)
        & (constraint_data["Monetary"] >= high_value_cutoff)
    ),
    "Constraint_Segment"
] = "Business_VIP_Outlier"

# Rule 2:
# Mark inactive low-frequency customers.
# They are inactive if Recency is high.
# They are low-frequency if Frequency is 2 or less.
# VIP customers are protected and not overwritten.

constraint_data.loc[
    (constraint_data["Recency"] >= inactive_cutoff)
    & (constraint_data["Frequency"] <= 2)
    & (constraint_data["Constraint_Segment"] != "Business_VIP_Outlier"),
    "Constraint_Segment"
] = "Business_Inactive_Low_Frequency"

# Create final business summary by segment.

constraint_summary = (
    constraint_data
    .groupby("Constraint_Segment")
    .agg(
        Customers=("CustomerID", "count"),
        Avg_Recency=("Recency", "mean"),
        Avg_Frequency=("Frequency", "mean"),
        Avg_Monetary=("Monetary", "mean"),
        Total_Monetary=("Monetary", "sum")
    )
    .reset_index()
)

# Add customer share and revenue share.
constraint_summary["Customer_Share_%"] = (
    constraint_summary["Customers"] / len(constraint_data) * 100
)

constraint_summary["Revenue_Share_%"] = (
    constraint_summary["Total_Monetary"] / constraint_data["Monetary"].sum() * 100
)

constraint_summary = constraint_summary.round(2)

# Save customer-level labels and summary table.
constraint_data.to_csv(
    OUTPUT_DIR / "constraint_based_segments.csv",
    index=False
)

constraint_summary.to_csv(
    OUTPUT_DIR / "constraint_based_summary.csv",
    index=False
)

constraint_summary
```

```python
# ------------------------------------------------------------
# 14.2 Draw constraint-based segment sizes
# ------------------------------------------------------------
# Goal:
# Show how many customers are in each final business segment.

# Sort segments so the largest segment appears at the top.
segment_chart_data = (
    constraint_summary
    .sort_values("Customers", ascending=True)
    .copy()
)

plt.figure(figsize=(10, 6))

bars = plt.barh(
    segment_chart_data["Constraint_Segment"],
    segment_chart_data["Customers"],
    alpha=0.80
)

plt.title("Constraint-Based Customer Segments", fontsize=14, weight="bold")
plt.xlabel("Number of customers")
plt.ylabel("Business segment")
plt.grid(axis="x", alpha=0.2)

# Add customer count and customer share next to each bar.
for bar, customers, share in zip(
    bars,
    segment_chart_data["Customers"],
    segment_chart_data["Customer_Share_%"]
):
    plt.text(
        bar.get_width(),
        bar.get_y() + bar.get_height() / 2,
        f"  {int(customers):,} customers ({share:.1f}%)",
        va="center",
        fontsize=9
    )

plt.tight_layout()
plt.show()
```
